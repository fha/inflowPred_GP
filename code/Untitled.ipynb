```python
import numpy as np
from sklearn import gaussian_process
from matplotlib import pylab as plt
def f(x):
    return np.sin(x)
X = np.atleast_2d([ 2.,6.,10.]).T
y = f(X).ravel()
x = np.atleast_2d(np.linspace(-50, 20, 1000)).T
gp = gaussian_process.GaussianProcess(theta0=1, thetaL=1e-2, thetaU=0.6)
gp.fit(X, y)  


y_pred, sigma2_pred = gp.predict(x, eval_MSE=True)

fig = pl.figure(figsize=(10, 10))

pl.plot(X, y, '*', label=u'$f(x) =\sin(x)$')
#pl.errorbar(x, y_pred, sigma2_pred, fmt='r.', markersize=10, label=u'Observations')
pl.plot(x, y_pred, 'b-', label=u'Prediction')
pl.fill(np.concatenate([x, x[::-1]]),
        np.concatenate([y_pred - 1.9600 * sigma2_pred,
                       (y_pred + 1.9600 * sigma2_pred)[::-1]]),
        alpha=.5, fc='g', ec='None', label='95% confidence interval')
pl.xlabel('$x$')
pl.ylabel('$f(x)$')

pl.xlim(-50,12)
pl.legend(loc='upper right')

pl.show()
```
